## Data analysis semeval visual WSD

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Load data

```python
with open('./data/test.data.v1.1.gold/en.test.data.v1.1.txt') as f:
    test_data_en = f.read().splitlines()

with open('./data/test.data.v1.1.gold/en.test.gold.v1.1.txt') as f:
    test_data_gold_en = f.read().splitlines()

with open('./data/semeval-2023-task-1-V-WSD-train-v1/train_v1/train.data.v1.txt') as f:
    train_data_en = f.read().splitlines()

with open('./data/semeval-2023-task-1-V-WSD-train-v1/train_v1/train.gold.v1.txt') as f:
    train_data_gold_en = f.read().splitlines()

with open('./data/semeval-2023-task-1-V-WSD-train-v1/trial_v1/trial.data.v1.txt') as f:
    trial_data_en = f.read().splitlines()

with open('./data/semeval-2023-task-1-V-WSD-train-v1/trial_v1/trial.gold.v1.txt') as f:
    trial_data_gold_en = f.read().splitlines()
```

```python
print(len(test_data_en))
print(len(test_data_gold_en))

print(len(train_data_en))
print(len(train_data_gold_en))

print(len(trial_data_en))
print(len(trial_data_gold_en))
```

```python
def read_texts(global_words, global_images, list_name):
    for l in list_name:
        line_split = l.split()
        words = []
        images = []
        for i in line_split:
            if "jpg" in i or "png" in i:
                images.append(i)
            else:
                words.append(i)
        global_words.append(words)
        global_images.append(images)
```

```python
test_words = []
test_images = []

read_texts(test_words, test_images, test_data_en)

train_words = []
train_images = []

read_texts(train_words, train_images, train_data_en)

trial_words = []
trial_images = []

read_texts(trial_words, trial_images, trial_data_en)
```

```python
def show_task(index, img_path, data_list_words, data_list_images, gold_list):
    fig = plt.figure(figsize=(20, 8))
    columns = 10
    rows = 1
    for i in range(1, columns*rows +1):
        jpg_path = data_list_images[index][i-1]
        img = mpimg.imread(img_path + jpg_path)
        fig.add_subplot(rows, columns, i)
        plt.imshow(img)
    plt.show()

    print(data_list_words[index])
    print(data_list_images[index].index(gold_list[index]))
```

### Test data

```python
show_task(0, "./data/test_images_resized/", test_words, test_images, test_data_gold_en)
```

### Training data

```python
show_task(2, "./data/semeval-2023-task-1-V-WSD-train-v1/train_v1/train_images_v1/", train_words, train_images, train_data_gold_en)
```
